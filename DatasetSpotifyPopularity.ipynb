```python
df_limpo = df.drop_duplicates(subset=['track_name'])

# Ordenar as músicas pelo nível de popularidade e pegar as 50 mais populares
top_50_musicas = df_limpo.sort_values(by='popularity', ascending=False).head(50)

# Criar o gráfico de barras horizontal
plt.figure(figsize=(12, 15))
plt.barh(top_50_musicas['track_name'] + ' - ' + top_50_musicas['track_genre'], top_50_musicas['popularity'], color='skyblue')
plt.xlabel('Popularidade')
plt.ylabel('Música - Gênero')
plt.title('Top 50 Músicas Mais Populares de 2023')
plt.gca().invert_yaxis()  # Inverter o eixo y para a música mais popular ficar no topo
plt.yticks(fontsize=8)  # Tamanho da fonte para o eixo y
plt.tight_layout()
plt.show()
```

```python
genero_frequencia = top_50_musicas['track_genre'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
genero_frequencia.plot(kind='bar', color='skyblue')
plt.xlabel('Gênero Musical')
plt.ylabel('Número de Aparições')
plt.title('Gêneros Musicais nas Top 50 Músicas Mais Populares de 2023')
plt.xticks(rotation=45, ha='right')  # Rotacionar os rótulos do eixo x para facilitar a leitura
plt.tight_layout()
plt.show()
```

```python
# artistas com mais de 100 tracks no dataset
df['popular_artist'] = df['artists'].map(df['artists'].value_counts()>100)
pop_arts  = df.groupby(['artists', 'popular_artist'])['popularity'].mean().sort_values(ascending=False).reset_index(1)
df_pop_arts = pop_arts.loc[pop_arts['popular_artist'] == True,['popularity']]
fig, ax = plt.subplots(figsize = (12, 10))
lead_artists = df_pop_arts.groupby('artists')['popularity'].mean().sort_values(ascending=False).head(50)
ax = sns.barplot(x=lead_artists.values, y=lead_artists.index, palette="Blues", orient="h", edgecolor='black', ax=ax)
ax.set_xlabel('Media de Popularidade', c='b', fontsize=12)
ax.set_ylabel('Artista', c='b', fontsize=12)
ax.set_title('Os 50 artistas mais populares no dataset', c='b', fontsize=14, weight = 'bold')

plt.show()
```

```python
df['duration_minutes'] = df['duration_ms'] / 60000

# Plotar o gráfico de dispersão
plt.figure(figsize=(10, 6))
plt.scatter(df['duration_minutes'], df['popularity'], alpha=0.5)
plt.title('Relação entre Popularidade e Duração das Músicas')
plt.xlabel('Duração (minutos)')
plt.ylabel('Popularidade')
plt.grid(True)

# Definir os intervalos e rótulos do eixo x
plt.xticks([0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80])
plt.xlim(0, 80)  # Limitar o eixo x até 80 minutos

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
ax = sns.barplot(x=['No Explicit', 'Explicit'],
                 y=(df['explicit'].value_counts(normalize=True) * 100),
                 palette='Spectral')
ax.set_ylabel('Porcentagem', fontsize=12, weight='bold')
ax.set_ylim(0, 100)
ax.set_yticks(range(0, 101, 10))
ax.set_yticklabels([f'{i}%' for i in range(0, 101, 10)], fontsize=12)

for p, value in zip(ax.patches, df['explicit'].value_counts()):
    percentage = f'{p.get_height():.2f}%'
    total = f'({value})'
    ax.annotate(f'{percentage}\n{total}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.suptitle('Ocorrência de Músicas Explícitas', fontsize=14, c='r')

plt.show()
```

```python
exp_data = sns.catplot(data=df, kind="bar",
                        x="explicit", y="popularity",
                        ci="sd", palette='Accent',
                        height=6)
exp_data.fig.suptitle('Popularity vs. Explicit', y=1.05, c='r')
exp_data.set_axis_labels("", "Popularity")
exp_data.set_xticklabels(['No Explicit\nContent', 'Explicit\nContent'])
plt.show()
```
